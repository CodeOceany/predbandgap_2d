```python
#导入库
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.preprocessing import scale
from sklearn.cross_validation import train_test_split
from sklearn.linear_model import Ridge
from sklearn.preprocessing import PolynomialFeatures
from sklearn.datasets import load_iris,load_boston
from sklearn.linear_model import BayesianRidge,LinearRegression,ElasticNet
from sklearn.svm import SVR
from sklearn.ensemble.gradient_boosting import GradientBoostingRegressor
from sklearn.model_selection import cross_val_score
from sklearn.metrics import explained_variance_score,mean_absolute_error,mean_squared_error,r2_score
```

```python
#导入数据
dataset = pd.read_csv('bandgap_MX2.csv',sep = ',',encoding = 'utf-8')
x=dataset.iloc[:,1:12]
y=dataset.iloc[:,12]
x=x - np.mean(x,axis=0)
print(x)
print(y)
```

```python
#数据标准化
data_standard = scale(dataset,with_mean=True,with_std=True)
print(dataset)
print(data_standard)
```

```python
#用view_model和model_worth来平谷模型,用build_model函数针对给定的数据构建岭回归模型
def build_model(x,y):
    model = Ridge(normalize=True,alpha=0.015)
    model.fit(x,y)
    return model
```

```python
def view_model(model):
    print("\n Model coeffiecients")
    print("===================\n")
    for i,coef in enumerate(model.coef_):
        print("\tCoefficient %d %0.3f"%(i+1,coef))
    print("\n\tIntercept %0.3f"%(model.intercept_))        
```

```python

def model_worth(true_y,predicted_y):
    print("\tMean squared error = %0.2f"%(mean_squared_error(true_y,predicted_y)))
    return mean_squared_error(true_y,predicted_y)
```

```python
if __name__ == "__main__":
    #将数据集分为训练集，dev集和测试集
    x_train,x_test_all,y_train,y_test_all=train_test_split(x,y,test_size = 0.3,random_state=10)
    x_dev,x_test,y_dev,y_test=train_test_split(x_test_all,y_test_all,test_size=0.5,random_state=10)
    
    #建立模型
    choosen_model=build_model(x_train,y_train)
    predicted_y=choosen_model.predict(x_train)
    print("\n Model performance in training set \n")
    mse=model_worth(y_train,predicted_y)
    view_model(choosen_model)
    
    #将模型用到dev数据集上
    predicted_y=choosen_model.predict(x_dev)
    print("\n Model performance in DEV set \n")
    model_worth(y_dev,predicted_y)
    
    #应用到测试集上
    predicted_y = choosen_model.predict(x_test)
    print("\n Model performance in test set \n")
    model_worth(y_test,predicted_y)
```
