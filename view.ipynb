```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

dataset = pd.read_csv('bandgap_MX2.csv')
x=dataset.iloc[:,1:12]
y=dataset.iloc[:,12]

sns.set(style='whitegrid',context='notebook')
cols=['Volume','Mass','Energy','Heat_of_formation','Electronegativity_A','Electronegativity_B','First_Ionization_energy_A','First_Ionization_energy_B','Radius_A','Radius_B','PBE','G0W0'
]
sns.pairplot(dataset[cols],size=2.5)
plt.show
```
